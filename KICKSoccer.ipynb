# Validation of Subcomponent Columns Against Their Totals

```python
# Check that each subcomponent does not exceed its total
checks_subcomponents = [
    ("shotsOnTarget", "totalShots"),
    ("penaltyKickGoals", "penaltyKickShots"),
    ("accuratePasses", "totalPasses"),
    ("accurateCrosses", "totalCrosses"),
    ("accurateLongBalls", "totalLongBalls"),
]
 
# For each pair (subcomponent, total)
for sub_col, total_col in checks_subcomponents:
    if sub_col in teamStats.columns and total_col in teamStats.columns:
        # Identify cases where the subcomponent is greater than the total
        subcomponent_issue = teamStats[teamStats[sub_col] > teamStats[total_col]]
        if not subcomponent_issue.empty:
            print(f"\nAnomaly: {sub_col} > {total_col} (count={len(subcomponent_issue)})")
            display(subcomponent_issue[['teamId', sub_col, total_col]].head())
        else:
            print(f"OK: no anomalies for {sub_col} vs {total_col} (subcomponents are consistent)")
```

# Validating Date Consistency in fixtures.csv

```python
from datetime import datetime

#  Check date consistency in fixtures.csv
if 'date' in fixtures.columns:
    
    def parse_date(date_str):
        """Convert a date string to a datetime object or return None if invalid."""
        formats = [
            "%Y-%m-%d", 
            "%Y-%m-%d %H:%M:%S", 
            "%d/%m/%Y", 
            "%m/%d/%Y", 
            "%Y-%m-%dT%H:%M:%S"
        ]
        for fmt in formats:
            try:
                return datetime.strptime(str(date_str), fmt)
            except ValueError:
                continue
        return None

    # Apply parsing
    parsed_dates = fixtures['date'].apply(parse_date)
    invalid_dates = parsed_dates.isna()

    # --- Check invalid/unparseable dates ---
    if invalid_dates.any():
        print(f"\n Anomaly: Invalid or unparseable dates (count={invalid_dates.sum()})")
        display(fixtures[invalid_dates][['eventId', 'date']].head())
    else:
        print("OK: All dates are parseable.")

    # --- Check date range validity ---
    valid_date_mask = ~invalid_dates
    date_min, date_max = datetime(2010, 1, 1), datetime(2030, 12, 31)

    out_of_range = fixtures[valid_date_mask][
        (parsed_dates < date_min) | (parsed_dates > date_max)
    ]

    if not out_of_range.empty:
        print(f"\n Warning: Dates out of the valid range [2010..2030] (count={len(out_of_range)})")
        display(out_of_range[['eventId', 'date']].head())
    else:
        print(" OK: No dates outside the range [2010..2030].")

else:
    print(" Warning: The column 'date' is missing from fixtures.csv.")

print("\n Checks completed for fixtures.csv")
```

## Validation of Subcomponent Columns Against Their Totals

```python
# Check that each subcomponent does not exceed its total
checks_subcomponents = [
    ("shotsOnTarget", "totalShots"),
    ("penaltyKickGoals", "penaltyKickShots"),
    ("accuratePasses", "totalPasses"),
    ("accurateCrosses", "totalCrosses"),
    ("accurateLongBalls", "totalLongBalls"),
]
 
# For each pair (subcomponent, total)
for sub_col, total_col in checks_subcomponents:
    if sub_col in teamStats.columns and total_col in teamStats.columns:
        # Identify cases where the subcomponent is greater than the total
        subcomponent_issue = teamStats[teamStats[sub_col] > teamStats[total_col]]
        if not subcomponent_issue.empty:
            print(f"\nAnomaly: {sub_col} > {total_col} (count={len(subcomponent_issue)})")
            display(subcomponent_issue[['teamId', sub_col, total_col]].head())
        else:
            print(f"OK: no anomalies for {sub_col} vs {total_col} (subcomponents are consistent)")
```

## Validating Ratio Columns Against Their Computed Values

```python
# List of ratio checks and error tolerance
ratio_checks = [
    ("shotPct",        "shotsOnTarget",    "totalShots"),
    ("passPct",        "accuratePasses",   "totalPasses"),
    ("crossPct",       "accurateCrosses",  "totalCrosses"),
    ("longballPct",    "accurateLongBalls","totalLongBalls")
]
tolerance = 0.1  # Acceptable margin of error
 
# For each ratio check
for ratio_col, sub_col, total_col in ratio_checks:
    if all(c in teamStats.columns for c in [ratio_col, sub_col, total_col]):
        mask = (teamStats[total_col] > 0)  # Avoid division by zero
        # Calculate the discrepancy between the displayed ratio and the computed ratio
        ratio_issue = teamStats[mask & (
            (teamStats[ratio_col] - (teamStats[sub_col] / teamStats[total_col])).abs() > tolerance
        )]
        if not ratio_issue.empty:
            print(f"\nAnomaly: {ratio_col} inconsistent with {sub_col}/{total_col} (count={len(ratio_issue)})")
            display(ratio_issue[[ratio_col, sub_col, total_col, "teamId"]].head())
        else:
            print(f"OK: no anomalies for {ratio_col} (ratio {sub_col}/{total_col}) within tolerance margin")
 
print("\n=== Checks completed. ===")
```

## Creating Performance Metrics and Comparison Features

```python
# Shot conversion rate (conversion) for the home team
import numpy as np
merged['home_shotConversion'] = np.where(
    merged['home_totalShots'] > 0,
    merged['home_shotsOnTarget'] / merged['home_totalShots'],
    0
)
 
# Shot conversion rate for the away team
merged['away_shotConversion'] = np.where(
    merged['away_totalShots'] > 0,
    merged['away_shotsOnTarget'] / merged['away_totalShots'],
    0
)
 
# Goal efficiency (identical to shot conversion rate)
merged['home_goalEfficiency'] = merged['home_shotConversion']
merged['away_goalEfficiency'] = merged['away_shotConversion']
 
# Ensure 'clean_sheet' exists, otherwise create it with 0
if 'home_clean_sheet' not in merged.columns:
    merged['home_clean_sheet'] = 0
else:
    merged['home_clean_sheet'] = merged['home_clean_sheet'].fillna(0)
 
if 'away_clean_sheet' not in merged.columns:
    merged['away_clean_sheet'] = 0
else:
    merged['away_clean_sheet'] = merged['away_clean_sheet'].fillna(0)
 
# Composite performance index (50% conversion, 50% clean sheet)
merged['home_performance_index'] = (
    0.5 * merged['home_shotConversion'] +
    0.5 * merged['home_clean_sheet']
)
merged['away_performance_index'] = (
    0.5 * merged['away_shotConversion'] +
    0.5 * merged['away_clean_sheet']
)
 
# Calculate differences between teams
merged['diff_shotConversion'] = merged['home_shotConversion'] - merged['away_shotConversion']
merged['diff_goalEfficiency'] = merged['home_goalEfficiency'] - merged['away_goalEfficiency']
merged['diff_performance_index'] = merged['home_performance_index'] - merged['away_performance_index']
 
# Points difference if available
if 'home_points' in merged.columns and 'away_points' in merged.columns:
    merged['diff_points'] = merged['home_points'] - merged['away_points']
```

##  Creating Match Statistics Differences and Penalty Efficiency

```python
# Difference in possession percentage between home and away teams
if 'home_possessionPct' in merged.columns and 'away_possessionPct' in merged.columns:
    merged['diff_possessionPct'] = merged['home_possessionPct'] - merged['away_possessionPct']
 
# Difference in fouls committed
if 'home_foulsCommitted' in merged.columns and 'away_foulsCommitted' in merged.columns:
    merged['diff_foulsCommitted'] = merged['home_foulsCommitted'] - merged['away_foulsCommitted']
 
# Difference in yellow cards
if 'home_yellowCards' in merged.columns and 'away_yellowCards' in merged.columns:
    merged['diff_yellowCards'] = merged['home_yellowCards'] - merged['away_yellowCards']
 
# Difference in red cards
if 'home_redCards' in merged.columns and 'away_redCards' in merged.columns:
    merged['diff_redCards'] = merged['home_redCards'] - merged['away_redCards']
 
# Difference in blocked shots
if 'home_blockedShots' in merged.columns and 'away_blockedShots' in merged.columns:
    merged['diff_blockedShots'] = merged['home_blockedShots'] - merged['away_blockedShots']
 
# Difference in effective clearances
if 'home_effectiveClearance' in merged.columns and 'away_effectiveClearance' in merged.columns:
    merged['diff_effectiveClearance'] = merged['home_effectiveClearance'] - merged['away_effectiveClearance']
 
# Penalty efficiency for the home team
if 'home_penaltyKickGoals' in merged.columns and 'home_penaltyKickShots' in merged.columns:
    merged['home_penaltyEfficiency'] = np.where(
        merged['home_penaltyKickShots'] > 0,
        merged['home_penaltyKickGoals'] / merged['home_penaltyKickShots'],
        0
    )
 
# Penalty efficiency for the away team
if 'away_penaltyKickGoals' in merged.columns and 'away_penaltyKickShots' in merged.columns:
    merged['away_penaltyEfficiency'] = np.where(
        merged['away_penaltyKickShots'] > 0,
        merged['away_penaltyKickGoals'] / merged['away_penaltyKickShots'],
        0
    )
```
